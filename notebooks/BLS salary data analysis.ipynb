```python
import pandas as pd
import requests
import json
import prettytable
```

```python
headers = {'Content-type': 'application/json'}
cpi_series_id = "CUUR0000SA0"
cpi_info = json.dumps({"seriesid":[cpi_series_id],"startyear":"2001","endyear":"2021"})
response = requests.post("https://api.bls.gov/publicAPI/v2/timeseries/data/",data=cpi_info,headers=headers)
cpi_json_data = json.loads(response.text)
```

```python
# print(json.dumps(json_data, indent=4, sort_keys=False))
response_results = []
for series in cpi_json_data['Results']['series']:    
    series_id = series['seriesID']
    for item in series['data']:
        entity={}
        entity['year'] = item['year']
        entity['period'] = item['period']
        entity['value'] = item['value']
        response_results.append(entity)
cpi_dataset = pd.DataFrame(response_results)
cpi_dataset.to_csv("../datasets/BLS_CPI.csv")
```
